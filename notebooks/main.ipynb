```python
import sys
sys.path.append('../')

import pandas as pd

import src.my_estimator as me
import src.my_transformers as mt
```

```python
df = pd.read_csv('../Test_Dataset/Training_DataSet.csv', header=0)
df.head()
```

```python
df_test = pd.read_csv('../Test_Dataset/Test_Dataset.csv', header=0)
df_test.head()
```

```python
df = mt.MainCleanerTransformer().fit_transform(df)
df = mt.NanCleanerTransformer().fit_transform(df)

df_test = mt.MainCleanerTransformer().fit_transform(df_test)
df_test = mt.NanCleanerTransformer().fit_transform(df_test)

df_xt5, df_wk2 = mt.XT5_WK2SplitterTransformer().fit_transform(df)
df_test_xt5, df_test_wk2 = mt.XT5_WK2SplitterTransformer().fit_transform(df_test)
df_xt5.shape, df_wk2.shape, df_test_xt5.shape, df_test_wk2.shape  
```

```python
ft = mt.FeatureTransformer()

df_xt5 = ft.transform(df_xt5)
df_wk2 = ft.transform(df_wk2)
df_test_xt5 = ft.transform(df_test_xt5)
df_test_wk2 = ft.transform(df_test_wk2)

df_xt5.shape, df_wk2.shape, df_test_xt5.shape, df_test_wk2.shape 
```

```python
te_xt5 = me.TrimEstimator()
te_xt5.fit(X=df_xt5, y=df_xt5[['Vehicle_Trim']])
df_test_xt5[['Vehicle_Trim']] = te_xt5.predict(X=df_test_xt5)
df_test_xt5.head()
```

```python
te_wk2 = me.TrimEstimator()
te_wk2.fit(X=df_wk2, y=df_wk2[['Vehicle_Trim']])
df_test_wk2[['Vehicle_Trim']] = te_wk2.predict(X=df_test_wk2)
df_test_wk2.head()
```

```python
ht_xt5 = mt.HistoryTransformer()
ohet_xt5 = mt.OHETransformer()
df_xt5 = ht_xt5.fit_transform(df_xt5)
df_xt5 = ohet_xt5.fit_transform(df_xt5)

df_test_xt5 = ht_xt5.transform(df_test_xt5)
df_test_xt5 = ohet_xt5.transform(df_test_xt5)
df_xt5.shape, df_test_xt5.shape
```
